```python
x=df.drop('price',axis=1) # X and Y Seprated Method
y=df['price']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_sc = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_sc, y, test_size=0.2, random_state=11)
```

```python
x_sc_train = sc.fit_transform(x_train)
x_sc_test = sc.transform(x_test)
```

```python
print(x_sc_train.shape)
print(x_sc_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_sc_train,y_train)
```

```python
lr.coef_ # m value
```

```python
lr.intercept_ # c value
```

```python
lr.predict([x_sc_test[0]])
```

```python
lr.score(x_sc_train,y_train)
```

```python
y_pred=lr.predict(x_sc_test)
```

```python
y_test
```

```python
result=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(mse)  #RMSE
```

```python
r2_score(y_test,y_pred)
```
