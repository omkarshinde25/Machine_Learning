```python
from tkinter import *
import pickle
win=Tk()
win.title('Home price prediction system')
win.geometry('1350x800')
win.configure(bg='skyblue')
def disp():
    area=var.get()
    f=open('model','rb')
    lr=pickle.load(f)
    f.close()
    result=lr.predict([[int(area)]])
    print(result)
    result=round(result[0],2)
    l3.config(text=str(result)+' Rs')
    var.set('')


l1=Label(win,text='HOME PRICE PREDICTION SYSTEM',
         relief='ridge',bd=5,font=('times new roman',16,'bold'))

l1.place(x=350,y=50)

l2=Label(win,text='Enter the Area',bd=5,relief='ridge',width=20)
l2.place(x=350,y=150)

var=StringVar()
e1=Entry(win,bd=5,relief='ridge',width=40,textvariable=var)
e1.place(x=510,y=150)

b1=Button(win,text='Predict Price',width=20,command=disp)
b1.place(x=510,y=190)

l3=Label(win,text='price is here',height=5,
         width=20,font=('times new roman',16,'bold'))
l3.place(x=350,y=250)

win.mainloop()
```
